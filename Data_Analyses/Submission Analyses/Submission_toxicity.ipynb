```python
import pandas as pd
import csv
```

# Relationship

```python
df_16=pd.read_csv("relationships_16_final.csv",sep=";")
df_17=pd.read_csv("relationships_17_final.csv",sep=";")
df_18=pd.read_csv("relationships_18_final.csv",sep=";")
df_19=pd.read_csv("relationships_19_final.csv",sep=";")
df_20=pd.read_csv("relationships_20_final.csv",sep=";")
df_21=pd.read_csv("relationships_21_final.csv",sep=";")
df_22=pd.read_csv("relationships_22_final.csv",sep=",")
```

```python
df_16_post_id=pd.read_csv("Relationship_16_post_double.csv")
df_17_post_id=pd.read_csv("Relationship_17_post_double.csv")
df_18_post_id=pd.read_csv("Relationship_18_post_double.csv")
df_19_post_id=pd.read_csv("Relationship_19_post_double.csv")
df_20_post_id=pd.read_csv("Relationship_20_post_double.csv")
df_21_post_id=pd.read_csv("Relationship_21_post_double.csv")
df_22_post_id=pd.read_csv("Relationship_22_post_double.csv")
```

# 2016

```python
# Merge der DataFrames basierend auf der "link_id"-Spalte
merged_df = pd.merge(df_16, df_16_post_id, how='inner', left_on='link_id', right_on='rel')
# Gruppieren nach der "link_id" und Berechnen des Durchschnitts für die "Toxicity"-Spalte
result_df = merged_df.groupby('link_id')[['Toxicity',"Severe Toxicity",'Profanity','Identity Attack','Insult','Threat']].mean().reset_index()
#result_df = merged_df.groupby('link_id')['Severe Toxicity'].mean().reset_index()
#result_df = merged_df.groupby('link_id')['Profanity'].mean().reset_index()
#result_df = merged_df.groupby('link_id')['Identity Attack'].mean().reset_index()
#result_df = merged_df.groupby('link_id')['Insult'].mean().reset_index()
#result_df = merged_df.groupby('link_id')['Threat'].mean().reset_index()
result_df.to_csv('Relationship_16_submission_all.csv', index=False,sep=";",quoting=csv.QUOTE_ALL)
result_df
```
