```python
import requests
from bs4 import BeautifulSoup
import csv

# Vorübergehend SSL-Verifikation deaktivieren
requests.packages.urllib3.disable_warnings()

# Die URL der Website
base_url = "https://botrank.pastimes.eu/?sort=rank&page="

# Die maximale Anzahl der Seiten auf der Website (angepasst nach Bedarf)
max_pages = 373

# Eine Liste, um die Bot-Namen zu speichern
bot_names = []

# Schleife durch die Seiten
for page_num in range(1, max_pages + 1):
    url = base_url + str(page_num)
    response = requests.get(url, verify=False)  # SSL-Verifikation vorübergehend deaktivieren

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        # Die Bot-Namen befinden sich in den 'a'-Tags in der zweiten Spalte ('td')
        bot_name_elements = soup.select('table.table td:nth-of-type(2) a')

        for element in bot_name_elements:
            bot_name = element.text.strip()
            bot_names.append(bot_name)
        
        print(f"Seite {page_num} verarbeitet.")  # Ausgabe der aktuellen Seite
    else:
        print(f"Fehler beim Abrufen der Seite {url}")

# Pfad zur CSV-Datei angeben (angepasst auf deinen Dateinamen)
csv_file_path = '/Users/anthonyfernando/Desktop/Masterarbeit/Reddit_analyses/bot_names/bot_names.csv'  # Hier den Dateinamen statt des Verzeichnisses angeben

# CSV-Datei erstellen und Bot-Namen speichern
with open(csv_file_path, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["Bot Name"])  # Überschrift für die Spalte

    for bot_name in bot_names:
        writer.writerow([bot_name])

print(f"Bot-Namen wurden in '{csv_file_path}' gespeichert.")








```
